```python
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
```

## 1. Rerata dan simpangan baku dari data pengukuran

```python
x = np.array([9, 5, 25, 23, 10, 22, 8, 8, 21, 20], dtype=float)
mean_direct = np.mean(x)
second_moment = np.mean(x**2)
std_direct = np.sqrt(second_moment - mean_direct**2)
mean_direct, second_moment, std_direct
```

## 2. Histogram dan distribusi probabilitas

```python
bin_centers = np.array([6, 9, 12, 15, 18, 21, 24], dtype=float)
bin_edges = np.array([4.5, 7.5, 10.5, 13.5, 16.5, 19.5, 22.5, 25.5])
counts, _ = np.histogram(x, bins=bin_edges)
prob = counts / len(x)
list(zip(bin_centers, counts, prob))
```

```python
fig, ax1 = plt.subplots(figsize=(6.5, 3.8))
ax1.bar(bin_centers, counts, width=3, edgecolor="black", alpha=0.35)
ax1.set_xlabel(r"$x_j$")
ax1.set_ylabel(r"$N(x_j)$")
ax1.set_xticks(bin_centers)
ax1.set_ylim(0, 5.2)

ax2 = ax1.twinx()
ax2.set_ylim(0, 0.52)
ax2.set_yticks(np.linspace(0, 0.5, 6))
ax2.set_ylabel(r"$P(x_j)$")
plt.show()
```

```python
mean_from_prob = np.sum(bin_centers * prob)
second_moment_from_prob = np.sum(bin_centers**2 * prob)
std_from_prob = np.sqrt(second_moment_from_prob - mean_from_prob**2)
mean_from_prob, second_moment_from_prob, std_from_prob
```

## 3. Latihan A: histogram dan variansi dari distribusi probabilitas

```python
A = np.array([10, 13, 14, 14, 6, 8, 7, 9, 12, 14, 13, 11, 10, 7, 7], dtype=float)
mean_A = np.mean(A)
var_A = np.mean(A**2) - mean_A**2
mean_A, var_A
```

```python
centers_A = np.array([5.5, 7.5, 9.5, 11.5, 13.5])
edges_A = np.array([4.5, 6.5, 8.5, 10.5, 12.5, 14.5])
counts_A, _ = np.histogram(A, bins=edges_A)
prob_A = counts_A / len(A)
mean_A_prob = np.sum(centers_A * prob_A)
var_A_prob_1 = np.sum((centers_A - mean_A_prob)**2 * prob_A)
var_A_prob_2 = np.sum(centers_A**2 * prob_A) - mean_A_prob**2
counts_A, prob_A, mean_A_prob, var_A_prob_1, var_A_prob_2
```

```python
plt.figure(figsize=(5.8, 3.5))
plt.bar(centers_A, counts_A, width=2, edgecolor="black", alpha=0.35)
plt.xlabel(r"$x_j$")
plt.ylabel(r"$N(x_j)$")
plt.xticks(centers_A)
plt.ylim(0, 5.5)
plt.show()
```

## 4. Probabilitas gabungan dan kondisional

```python
B = [(3,4), (5,8), (4,4), (8,5), (3,5), (4,5), (5,8), (8,5), (8,4), (3,4), (3,8), (4,8)]
N = len(B)
x_vals = [p[0] for p in B]
y_vals = [p[1] for p in B]
Px = {k: v/N for k, v in Counter(x_vals).items()}
Py = {k: v/N for k, v in Counter(y_vals).items()}
P_joint_5_8 = Counter(B)[(5,8)]/N
P_cond_5_given_8 = Counter([x for x, y in B if y == 8])[5] / Counter(y_vals)[8]
Px, Py, P_joint_5_8, P_cond_5_given_8, P_cond_5_given_8 * Py[8]
```

## 5. Pemeriksaan simbolik untuk distribusi kontinu

```python
import sympy as sp
x_sym, alpha, beta, c = sp.symbols("x alpha beta c", positive=True)
normalization = sp.integrate(c * sp.exp(-alpha * x_sym), (x_sym, 0, sp.oo))
sp.solve(sp.Eq(normalization, 1), c)
```

```python
p_exp = alpha * sp.exp(-alpha * x_sym)
mean_exp = sp.integrate(x_sym * p_exp, (x_sym, 0, sp.oo))
second_exp = sp.integrate(x_sym**2 * p_exp, (x_sym, 0, sp.oo))
std_exp = sp.sqrt(second_exp - mean_exp**2)
mean_exp, second_exp, sp.simplify(std_exp)
```

```python
p_gauss = 1/(beta * sp.sqrt(sp.pi)) * sp.exp(-(x_sym-alpha)**2 / beta**2)
mean_gauss = sp.integrate(x_sym * p_gauss, (x_sym, -sp.oo, sp.oo))
second_gauss = sp.integrate(x_sym**2 * p_gauss, (x_sym, -sp.oo, sp.oo))
std_gauss = sp.sqrt(sp.simplify(second_gauss - mean_gauss**2))
sp.simplify(mean_gauss), sp.simplify(second_gauss), sp.simplify(std_gauss)
```
